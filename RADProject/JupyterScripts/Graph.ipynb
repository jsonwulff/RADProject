```python
import numpy as np
import matplotlib.pyplot as plt

data_range = [10, 20, 30, 40, 50, 60, 70, 80, 90]
line = 0
mse_data = np.loadtxt('mse_values.txt', delimiter=',')
chi_data = np.loadtxt('chi_values.txt', delimiter=',')
s_data = np.loadtxt('s_values.txt', delimiter=',')
m_data = np.loadtxt('m_values.txt', delimiter=',')
expectedValRes = 0
varians = 0
for i in range(0, 100):
    #print(i, chi_data[0][i])
    expectedValRes += chi_data[line][i]*1/100;

for i in range(0,100):
    varians += (chi_data[line][i] - expectedValRes)**2; 
    
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(data_range,m_data[line], color='r', label = 'Median estimate values')
ax.set_xlabel('Interval')
ax.set_ylabel('S values')
ax.set_title('Hashing results, m = 10')
plt.ylim(2000000,5500000)

print(varians/100)
print(expectedValRes)
#print(m_data[line])
#print(s_data[line])
plt.hlines(s_data[line],0,100,color='b', label ='Actual S value')
plt.plot(chi_data[line], color='g', label='Chi data')
plt.legend()
```
